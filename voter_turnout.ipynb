```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_turnout = pd.read_csv("../voter_turnout.csv")
df_turnout["turnout_vap"] = df_turnout["turnout_vap"].apply(lambda x: float(x[:-1]))
df_turnout["turnout_vep"] = df_turnout["turnout_vep"].apply(lambda x: float(x[:-1]))

df_turnout
```

```python
plt.plot(df_turnout["year"], df_turnout["turnout_vap"], "-s")
plt.ylabel("Turnout in percent of voting age population")
plt.xlabel("year")
plt.grid()
```
